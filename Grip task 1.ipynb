```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("http://bit.ly/w-data")
data
```

```python
x=data.iloc[:,:1].values
x
```

```python
y=data.iloc[:,1].values
y
```

```python
#seprating down the features(inputs)and labels(outputs) using sklearn library

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
x_test
```

```python
y_test
```

```python
x_train
```

```python
y_train
```

```python
#training data using Linear Regression
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
y_pred
```

```python
#visualizing the data by plotting the graph for the data (using matplotlib)
```

```python
plt.scatter(x_train,y_train,color='r')
plt.plot(x_train,reg.predict(x_train))
plt.show()
```

```python
#predicting score of an student who studies 9.25hours/dayÂ¶
hours=([9.25])
pred_score=reg.predict([hours])
print("study hours:",("hours"))
print("predicted score:",(pred_score))
```
